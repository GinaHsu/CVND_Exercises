# HSV Color Space

### Import resources and display image

## Color Selection
   To select the most accurate color boundaries, it's often useful to use a [color picker](http://www.w3schools.com/colors/colors_picker.asp) and choose the color boundaries that define the region you want to select! 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline

# Read in the image
image = cv2.imread('images/1_1_test_blue.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
print('This image is:', type(image_copy), 
      ' with dimensions:', image_copy.shape)
```

### Plot color channels

```python
# RGB channels
#[:,:,] take all the x and y coordinates of the image pixcels
r = image_copy[:,:,0] #the zero index which is the red value for each pixel
g = image_copy[:,:,1] #the 2nd index which is the green value for each pixel
b = image_copy[:,:,2] #the 3rd index which is the blue value for each pixel

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Red') 
ax1.imshow(r, cmap='gray')#the pink balloons have high values for red

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')#medium high values for blue
```

```python
# Convert from RGB to HSV
hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

### Define blue and HSV selection thresholds

```python
# Define our color selection criteria in RGB values (0-255)
#the thresholds allow some high values of red and medium values of blue
lower_blue = np.array([0,0,120]) 
upper_blue = np.array([230,230,255])
```

### Mask the image 

```python
# Define the masked area in RGB space
mask_rgb = cv2.inRange(image_copy, lower_blue, upper_blue)

# mask the image
masked_image = np.copy(image_copy)
masked_image[mask_rgb==0] = [0,0,0]
```

```python
# Now try HSV!

# Define our color selection criteria in HSV values (0-180)
#(hue, saturation, value)
#the thresholds allow a narrow high range for the hue channel from 160-180 
lower_hue= np.array([100,100,170]) 
upper_hue = np.array([200,255,255])

# Define the masked area in HSV space
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

# mask the image
masked_image_hsv = np.copy(image_copy)
masked_image_hsv[mask_hsv==0] = [0,0,0]

# Vizualize the 2 masks
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('The image of RGB MASK')
ax1.imshow(masked_image)

ax2.set_title('The image of HSV MASK')
ax2.imshow(masked_image_hsv)
```

### Mask and add a background image

```python
# Load in a background image, and convert it to RGB 
background_image = cv2.imread('images/space_background.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

# Crop it to the right size (960x1280)
crop_background = background_image[0:960, 0:1280]

# Mask the cropped background so that the pizza area is blocked
crop_background[mask_hsv != 0] = [0, 0, 0]

# Display the background
plt.imshow(crop_background)
```

### Create the complete images

```python
complete_img = crop_background + masked_image
complete_img_hsv = crop_background + masked_image_hsv

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,10))

ax1.set_title('The completed image of RGB MASK')
ax1.imshow(complete_img)

ax2.set_title('The completed image of HSV MASK')
ax2.imshow(complete_img_hsv)
```
